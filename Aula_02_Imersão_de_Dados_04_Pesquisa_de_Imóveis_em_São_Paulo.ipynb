Como estou tentando usar uma função que é usanda em str em uma Series inteira do Data Frame, eu preciso informar que essa função é para str.

Adicionando no Data Frame essas colunas obtidas com o split.

```python
data_base[['Moeda', 'Valor_Imovel', 'Tipo_de_Contrato']] = data_base['Valor'].str.split(expand = True)
data_base
```

Vou querer apenas aquelas linhas do Data Frame cujo o Tipo_de_Contrato é igual a null.

```python
data_base[data_base['Tipo_de_Contrato'].isnull()]
```

```python
data_base[data_base['Tipo_de_Contrato'].isnull()]['Tipo_de_Contrato'].unique()
```

Ao invés de salvar no Data Frame original, é aconselhavel criar uma outra variável para armazenar os dados filtrados.

```python
filtered_data = data_base[data_base['Tipo_de_Contrato'].isnull()]
filtered_data
```

Transformando os dados de Valor_do_Imovel para o tipo float.

```python
filtered_data['Valor_Imovel'].str.replace('.', '').astype(float)
```

```python
filtered_data['Valor_Imovel_float'] = filtered_data['Valor_Imovel'].str.replace('.', '').astype(float)
filtered_data
```

#### Desafio 1: Criar uma escala de valores em milhões.

```python
filtered_data['Valor_Imovel_float'].apply(lambda x: x/1000000)
```

Também é possível fazer com a função transform

```python
filtered_data['Valor_Imovel_float'].transform(lambda x: x/1000000)
```

```python
filtered_data['Valor_Imovel_float_milhao'] = filtered_data['Valor_Imovel_float'].apply(lambda x: x/1000000)

filtered_data[['Valor_Imovel', 'Valor_Imovel_float', 'Valor_Imovel_float_milhao']].sample(21)
```

#### Desafio 2: Deixar o gráfico do histograma, de valores, legível (alterar labels, cores, título, escala).

```python
plt.figure(figsize = (10, 5))
values_histogram = sns.histplot(data = filtered_data, x = 'Valor_Imovel_float_milhao', color = 'Orange', binwidth = 21, element = 'poly')
values_histogram.set(title = 'Relação Entre a Quantidade de Imóveis e o seu Valor em Milhões de Reais')
sns.set_style('whitegrid')
values_histogram.set_xlabel('Valor (R$)')
values_histogram.set_ylabel('Quantidade')
plt.show()
```

#### Desafio 3: Preço do metro quadrado por bairro e plotar em um gráfico ideal.

```python
mean_by_district = filtered_data[['Bairro', 'Valor_Imovel_float', 'Metragem']].groupby(['Bairro']).mean()
mean_by_district['Media_Preco_Metro_Por_Bairro'] = mean_by_district['Valor_Imovel_float'].div(mean_by_district['Metragem'])
mean_by_district = mean_by_district.sort_values(by = ['Media_Preco_Metro_Por_Bairro'], ascending = False)
mean_by_district
```

```python
plt.figure(figsize = (21, 10))
mean_by_district_lineplot = sns.lineplot(data = mean_by_district['Media_Preco_Metro_Por_Bairro'].head(10), linewidth = 1.5)
mean_by_district_lineplot.set(title = 'Média de Preço do Metro Quadrado por Bairro')
mean_by_district_lineplot.set_ylabel('Preço (R$)')
plt.show()
```

Outra maneira de obter a média de preço do metro:

```python
filtered_data['V/M'] = filtered_data['Valor_Imovel_float'] / filtered_data['Metragem']
filtered_data
```

```python
by_district = filtered_data[['Bairro', 'V/M']].groupby(['Bairro']).mean().sort_values(by = ['V/M'], ascending = False)
by_district
```
