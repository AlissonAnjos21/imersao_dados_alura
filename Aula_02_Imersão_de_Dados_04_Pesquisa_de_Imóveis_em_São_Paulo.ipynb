## Importando Arquivo

```python
import pandas as pd

path = 'https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv'
data_base = pd.read_csv(path)
```

Como estou tentando usar uma função que é usanda em str em uma Series inteira do Data Frame, eu preciso informar que essa função é para str.

```python
data_base['Valor'].str.split()
```

Consigo criar uma coluna para cada um dos elementos que foram separados por espaço.

```python
data_base['Valor'].str.split(expand = True)
```

Felizmente, na primeira coluna não existe nenhum número misturado com os $.

```python
data_base['Valor'].str.split(expand = True)[0].unique()
```

Felizmente, também não existe nenhum número misturado com essa terceira coluna

```python
data_base['Valor'].str.split(expand = True)[2].unique()
```

```python
len(data_base['Valor'].str.split(expand = True)[1].unique())
```

Adicionando no Data Frame essas colunas obtidas com o split.

```python
data_base[['Moeda', 'Valor_Imovel', 'Tipo_de_Contrato']] = data_base['Valor'].str.split(expand = True)
data_base
```

Vou querer apenas aquelas linhas do Data Frame cujo o Tipo_de_Contrato é igual a null.

```python
data_base[data_base['Tipo_de_Contrato'].isnull()]
```

```python
data_base[data_base['Tipo_de_Contrato'].isnull()]['Tipo_de_Contrato'].unique()
```

Ao invés de salvar no Data Frame original, é aconselhavel criar uma outra variável para armazenar os dados filtrados.

```python
filtered_data = data_base[data_base['Tipo_de_Contrato'].isnull()]
filtered_data
```

Transformando os dados de Valor_do_Imovel para o tipo float.

```python
filtered_data['Valor_Imovel'].str.replace('.', '').astype(float)
```

```python
filtered_data['Valor_Imovel_float'] = filtered_data['Valor_Imovel'].str.replace('.', '').astype(float)
filtered_data
```

```python
filtered_data.sample(21)
```

Perceba que o tipo de Valor_Imovel_float realmente é float.

```python
filtered_data.info()
```
