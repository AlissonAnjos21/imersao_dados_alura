#### Desafio 3: Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão.

```python
district_mean = data_base[['Bairro', 'Metragem']].groupby(['Bairro']).mean()
district_mean = district_mean.sort_values(by=['Metragem'], ascending = False)
district_mean
```

Gráfico dos 10 bairros com a maior média de metragem:

Percebe-se que a diferença entre a média de matragem do bairro Sítio Repreza, em comparação aos outros 9 maiores, é gigantescamente maior que a diferença que as outras maiores possuem entre si.

```python
district_mean.head(10).plot.pie(subplots = True, figsize = (10, 10))
```

Gráfico dos 10 bairros com a menor média de metragem:

Percebe-se que o bairro, Vila Guaca, com a menor média metragem é a que possui o maior grau de diferença entre os demais menores selecionados, menores esses que não possuem grande diferença de metragem média entre si.

```python
district_mean.tail(10).plot.pie(subplots = True, figsize = (10, 10))
```

```python
district_rooms = data_base.loc[:, ['Bairro', 'Quartos']].groupby(['Bairro']).mean()
district_rooms = round(district_rooms.sort_values(by=['Quartos'], ascending = False))
district_rooms
```

Os 10 bairros com o maior número médio de quartos:

Percebe-se que são poucos os bairros que possuem a média de quartos maior que 6, ao passo que, existe um número considerável de bairros com a média de 6 quartos.

```python
district_rooms.head(10).sort_values(by = ['Quartos'], ascending = True).plot.barh(figsize = (10, 10))
```
